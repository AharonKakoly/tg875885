```python
import numpy as np
import math
A='/home1/08289/tg875885/radial_to_rotating_flows/Aharon/HalosData/H29.dat'
DataOfHalos=np.loadtxt(A)
Snapnum=[]
Snapz=[]
Rvirs=[]
Masses=[]
logMasses=[]
print(DataOfHalos[175][0])
print(DataOfHalos[63][0])
for i in range (175,62,-1):
    Snapnum.append(DataOfHalos[i][0])
    Snapz.append(DataOfHalos[i][1])
    Rvirs.append(DataOfHalos[i][13]/(1+DataOfHalos[i][1]))
    Masses.append(DataOfHalos[i][5])
    logMasses.append(math.log10(DataOfHalos[i][5]))
    
print("")    
print(Snapnum)
print("")
print(Snapz)
print("")
print(Rvirs)
print("")
print(Masses)
print("")
print(logMasses)
```
