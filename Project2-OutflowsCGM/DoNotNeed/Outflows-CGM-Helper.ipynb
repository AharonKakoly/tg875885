```python
import numpy as np
import math
A='/home1/08289/tg875885/radial_to_rotating_flows/Aharon/HalosData/m12i.dat'
DataOfHalos=np.loadtxt(A)
Snapnum=[]
Snapz=[]
Rvirs=[]
Masses=[]
logMasses=[]
print(DataOfHalos[0][0])
print(DataOfHalos[386][0])
for i in range (386,-1,-1):
    Snapnum.append(DataOfHalos[i][0])
    Snapz.append(DataOfHalos[i][1])
    Rvirs.append(DataOfHalos[i][13]/(1+DataOfHalos[i][1]))
    Masses.append(DataOfHalos[i][5])
    logMasses.append(math.log10(DataOfHalos[i][5]))

Redshift0_025=[]
Snapshot0_025=[]
    
Redshift025_05=[]
Snapshot025_05=[]

Redshift05_075=[]
Snapshot05_075=[]

Redshift075_1=[]
Snapshot075_1=[]

Redshift1_125=[]
Snapshot1_125=[]

Redshift125_150=[]
Snapshot125_150=[]

for i in range(0,len(Snapz)):
    if(Snapz[i]>=0 and Snapz[i]<=0.25):
        Redshift0_025.append(Snapz[i])
        Snapshot0_025.append(Snapnum[i])
    
    if(Snapz[i]>=0.25 and Snapz[i]<=0.5):
        Redshift025_05.append(Snapz[i])
        Snapshot025_05.append(Snapnum[i])
        
    if(Snapz[i]>0.5 and Snapz[i]<=0.75):
        Redshift05_075.append(Snapz[i])
        Snapshot05_075.append(Snapnum[i])
        
    if(Snapz[i]>0.75 and Snapz[i]<=1):
        Redshift075_1.append(Snapz[i])
        Snapshot075_1.append(Snapnum[i])
        
    if(Snapz[i]>1 and Snapz[i]<=1.25):
        Redshift1_125.append(Snapz[i])
        Snapshot1_125.append(Snapnum[i])
        
    if(Snapz[i]>1.25 and Snapz[i]<=1.5):
        Redshift125_150.append(Snapz[i])
        Snapshot125_150.append(Snapnum[i])
        
print("Redshift 0 to 0.25")
print(Redshift0_025)
print(Snapshot0_025)
print()        
        
print("Redshift 0.25 to 0.5")
print(Redshift025_05)
print(Snapshot025_05)
print()

print("Redshift 0.5 to 0.75")
print(Redshift05_075)
print(Snapshot05_075)
print()

print("Redshift 0.75 to 1.0")
print(Redshift075_1)
print(Snapshot075_1)
print()

print("Redshift 1.0 to 1.25")
print(Redshift1_125)
print(Snapshot1_125)
print()

print("Redshift 1.25 to 1.5")
print(Redshift125_150)
print(Snapshot125_150)
print()
        
#print("")    
#print(Snapnum)
#print("")
#print(Snapz)
#print("")
#print(Rvirs)
#print("")
#print(Masses)
#print("")
#print(logMasses)
```
