```python
import numpy as np
import math
from numpy import load

#  For m12i_res7100
print("Simulation name: m12i_res7100")
data=load('/home1/08289/tg875885/radial_to_rotating_flows/Aharon/HalosMasses/m12i_mvirs.npz')
ListOfSnapshotNumbers=[77,102,142,214]
lst=data.files
ListOfKeys=[]
for item in lst:
    ListOfKeys.append(str(item))   
print("File keys are: ",ListOfKeys)
print("")

Zs=data['zs']
SnapshotNum=data['snapshotNum']
Mvirs=data['mvirs']
SignificantDigits=2

for i in ListOfSnapshotNumbers:
    indexx=np.array([np.where(SnapshotNum==i)])
    index=indexx[0][0][0]
    print("Snapshot number is: ",SnapshotNum[index])
    print("Snapshot index is: ",index)
    print("Snapshot Z is: ",Zs[index])
    print("Snapshot halo mass is: ",Mvirs[index])
    RoundedMass=round(Mvirs[index],SignificantDigits-int(math.floor(math.log10(abs(Mvirs[index]))))-1)
    SientificNotation="{:.2e}".format(RoundedMass)
    print("Snapshot halo rounded mass is: ",SientificNotation)

    print("")
```
